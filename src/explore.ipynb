### ESCALANDO
#### Guardando Datasets

```python
from sklearn.model_selection import train_test_split

predictoras = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age"]
target = 'Outcome'

x = data_diag[predictoras]
y = data_diag[target]

x_train, x_test, y_train, y_test = train_test_split ( x, y, test_size = 0.3, random_state = 7)

data_diag_train = pd.concat ([x_train, y_train], axis = 1)
data_diag_train.to_csv ("/workspaces/Decission_Tree_LeryBenavides/data/raw/DATASET_TRAIN.csv")

data_diag_test = pd.concat ([x_test, y_test], axis = 1)
data_diag_test.to_csv ("/workspaces/Decission_Tree_LeryBenavides/data/raw/DATASET_TEST.csv")
```

```python
from sklearn.preprocessing import StandardScaler
from pickle import dump

normalizer = StandardScaler()
normalizer.fit (x_train)

x_train_norm = normalizer.transform (x_train)
x_train_norm = pd.DataFrame (x_train_norm, index = x_train.index, columns = predictoras)

x_test_norm = normalizer.transform (x_test)
x_test_norm = pd.DataFrame (x_test_norm, index = x_test.index, columns = predictoras)

data_diag_train_norm = pd.concat ([x_train_norm, y_train], axis = 1)
data_diag_train_norm.to_csv ("/workspaces/Decission_Tree_LeryBenavides/data/raw/DATASET_TRAIN_NORM.csv")
data_diag_test_norm = pd.concat ([x_test_norm, y_test], axis = 1)
data_diag_test_norm.to_csv ("/workspaces/Decission_Tree_LeryBenavides/data/raw/DATASET_TEST_NORM.csv")

dump (normalizer, open ("/workspaces/Decission_Tree_LeryBenavides/models/normalizer.sav", "wb"))

data_diag_train_norm
```

### FEATURE SELECTION

```python
from sklearn.feature_selection import f_regression, SelectKBest

selector_model = SelectKBest (f_regression, k = 5)
selector_model.fit (x_train_norm, y_train)

best_variables = selector_model.get_support()

x_train_sel = pd.DataFrame (x_train_norm, columns = x_train_norm.columns[best_variables])
x_test_sel = pd.DataFrame (x_test_norm, columns = x_test_norm.columns[best_variables])

x_train_sel
```

## DECISSION TREE MODEL

```python
train_data = pd.read_csv ("/workspaces/Decission_Tree_LeryBenavides/data/processed/clean_train.csv")
test_data = pd.read_csv ("/workspaces/Decission_Tree_LeryBenavides/data/processed/clean_test.csv")

train_data.head()
```

```python
plt.figure (figsize = (12,6))

pd.plotting.parallel_coordinates (data_diag, "Outcome", color = ("#E58139", "#39E581", "#8139E5"))

plt.show()
```

```python
x_DT_train = train_data.drop(["Outcome"], axis = 1)
y_DT_train = train_data["Outcome"]

x_DT_test = test_data.drop(["Outcome"], axis = 1)
y_DT_test = test_data["Outcome"]
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier (random_state = 7)
model.fit (x_DT_train, y_DT_train)
```

```python
from sklearn import tree

fig = plt.figure (figsize = (15,15))

tree.plot_tree (model, feature_names = list(x_DT_train.columns), class_names = ["0", "1", "2"], filled = True)

plt.show()
```

```python
y_pred = model.predict(x_DT_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_DT_test, y_pred)
```

### OPTIMIZACIÃ“N

```python
from sklearn.model_selection import GridSearchCV

hyperparams = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

grid = GridSearchCV(model, hyperparams, scoring = "accuracy", cv = 10)
grid
```

```python
import warnings

with warnings.catch_warnings():
    warnings.simplefilter("ignore")

grid.fit (x_DT_train, y_DT_train)

grid.best_params_
```

```python
best_model = DecisionTreeClassifier (criterion = "entropy", min_samples_leaf = 2, min_samples_split = 10, random_state = 7)
best_model.fit (x_DT_train, y_DT_train)
```

```python
best_y_pred = best_model.predict (x_DT_test)
best_y_pred
```

```python
accuracy_score(y_DT_test, best_y_pred)
```
